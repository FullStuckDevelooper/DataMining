```
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.tree import export_graphviz
from IPython.display import Image
import pydotplus
df = pd.read_csv("bank.csv",delimiter=';')
```

```
df
```

```
df.drop(['contact','day','month','poutcome','job','education'] , axis=1, inplace=True)
```

```
df.info()
```

```
df.head()
```

```
martial_mapping_values = {"married": 1, "single": 0, "divorced":3}
df.replace({"marital":martial_mapping_values}, inplace=True)

yes_no_mapping_values = {"yes": 1, "no": 0}
df.replace({"default":yes_no_mapping_values}, inplace=True)
df.replace({"housing":yes_no_mapping_values}, inplace=True)
df.replace({"loan":yes_no_mapping_values}, inplace=True)
df.replace({"y":yes_no_mapping_values}, inplace=True)

```

```
df.head()
```

```
x = df.drop("y", axis=1)
```

```
y = df["y"]
```

```
parameters = {'max_depth':range(3,15)}

clf = GridSearchCV(DecisionTreeClassifier(), parameters)
clf.fit(x, y)
tree_model = clf.best_estimator_
```

```
print("Best score : ", clf.best_score_)
print("Best parameter: ", clf.best_params_)
```

```
tree_model
```

```
dot_data = export_graphviz(
    tree_model,
    out_file = None,
    feature_names = list(x.columns.values),
    class_names = ['Yes', 'No'],
    rounded = True,
    filled = True
)

graph = pydotplus.graph_from_dot_data(dot_data)

Image(graph.create_png())
```
